```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_hdf("data/train_mod.h5", low_memory = False)
df.head()
```

```python
df = df[df.passenger_count <= 7]
```

```python
df.drop(columns = ['index'], inplace = True)
df.reset_index(inplace = True, drop = True)
```

```python
df.head()
```

```python
df_copy = df.copy()
```

```python
from sklearn import preprocessing
scaler_mm = preprocessing.MinMaxScaler()
scaler_std = preprocessing.StandardScaler()
```

```python
mm_features = df_copy[['passenger_count', 'year']]
std_features = df_copy[['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']]
scaler_mm.fit(mm_features)
mm_features = scaler_mm.transform(mm_features)
```

```python
mm_features = pd.DataFrame(mm_features, columns = ['passenger_count', 'year'])
mm_features.head()
```

```python
scaler_std.fit(std_features)
std_features = scaler_std.transform(std_features)
std_features = pd.DataFrame(std_features, columns = ['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude'])
std_features.head()
```

```python
df[['passenger_count', 'year']] = mm_features
df[['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude']] = std_features
df.head()
```

```python
df.to_hdf(r'data/train_final.h5', 'data')
```
